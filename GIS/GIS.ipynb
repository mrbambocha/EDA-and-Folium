```python
One=sensor_data[['app_id','latitude','longitude','sensor_name']]
Two=Here[['id','latitude','longitude','brand']]
```

```python
Combined['ID'] = Combined[['app_id', 'sensor_name']].apply(lambda x: ','.join(x.fillna('').map(str)), axis=1)
Combined['ID'] = Combined['ID'].str.strip(',')
```

```python
def regioncolors(counter):
    if counter['sensor_name'] == 'Elsys ELT2':
        return 'darkred'
    elif counter['sensor_name'] == 'Elsys ERS CO2':
        return 'white'
    elif counter['sensor_name'] == 'ESP32':
        return 'red'
    elif counter['sensor_name'] == 'Arduino':
        return 'darkpurple'
    elif counter['sensor_name'] == 'The Things Products':
        return 'gray'
    elif counter['sensor_name'] == 'Laird':
        return 'black'
    elif counter['sensor_name'] == 'Gemtek':
        return 'cadetblue'
    elif counter['sensor_name'] == 'ESP based':
        return 'purple'
    elif counter['sensor_name'] == 'MikroTik':
        return 'lightred'
    elif counter['sensor_name'] == 'n-Fuse':
        return 'pink'
    elif counter['sensor_name'] == 'Raspberry Pi based':
        return 'red'
    elif counter['sensor_name'] == 'MatchX':
        return 'darkblue'
    elif counter['sensor_name'] == 'TTGO':
        return 'darkgreen'
    elif counter['sensor_name'] == 'The Things Industries':
        return 'orange' 
    elif counter['sensor_name'] == 'Tannak International':
        return 'lightgray'
    elif counter['sensor_name'] == 'VestiTech':
        return 'beige'
    elif counter['sensor_name'] == 'RAKwireless':
        return 'lightgreen'
    elif counter['sensor_name'] == 'Dragino':
        return 'blue'
    elif counter['sensor_name'] == 'Heltec':
        return 'lightblue'
    elif counter['sensor_name'] == 'MultiTech':
        return 'green'
    else:
        return 'darkblue'
train_df["color"] = train_df.apply(regioncolors, axis=1)
train_df.head()
```

```python
map3 = folium.Map(location=[59.481945, 17.925127], tiles='CartoDB positron', zoom_start=6,radius=4000)

#marker_cluster =  MarkerCluster().add_to(map3)

for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup='ID:'+train_df['ID'][point], icon=folium.Icon(color=train_df["color"][point], icon_color='white', icon='male', angle=0, prefix='fa')).add_to(map3)
map3
```

```python
import folium
import numpy as np
from geopy.distance import geodesic

points = locationlist
center = (np.mean([a for a,b in points]), np.mean([b for a,b in points]))
MAP = folium.Map(location=[59.481945, 17.925127], tiles='CartoDB positron', zoom_start=6,radius=50)
r = np.sin(5 *    np.linspace(0, 2*np.pi, 68))

lgd_txt = '<span style="r: {col};">{txt}</span>'
for p in points:
    folium.Marker(p).add_to(MAP)

for idx, r in enumerate( [0.2, 0.5,1,0.7,0.3]):  # color choice is limited
    print(r)
    fg = folium.FeatureGroup(name= lgd_txt.format( txt= str(r)+' radius', col=r ))
    
    p1 = folium.Circle(location = list(center),
                    radius = max([geodesic(center,point).meters for point in points])*r,
                    color = '#3186cc',
                    fill = True,
                    fill_color = '#1086cd')
    fg.add_child(p1)
#    fg.add_child(p2)
    MAP.add_child(fg)
    



#marker_cluster =  MarkerCluster().add_to(map3)

for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup='ID:'+train_df['ID'][point], icon=folium.Icon(color=train_df["color"][point], icon_color='white', icon='male', angle=0, prefix='fa')).add_to(MAP)
folium.map.LayerControl('topleft', collapsed= False).add_to(MAP)
MAP.save('sensorandAPI.html')
MAP
```
